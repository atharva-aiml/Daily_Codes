```python
def replacer(df):
    cat = list(df.columns[df.dtypes=="object"])
    for i in df.columns:
        if i in cat:
            df[i] = df[i].fillna(df[i].mode()[0])
        else:
            df[i] = df[i].fillna(df[i].mean())

print("Done!!")
```

```python
replacer(df)
```

```python
x_cat = df.select_dtypes(include="object")
x_cat.head()
```

```python
x_con = df.select_dtypes(exclude="object")
x_con.head()
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
```

```python
ohe = OneHotEncoder(handle_unknown="ignore", sparse_output=False).set_output(transform="pandas")
```

```python
x_cat_pre = ohe.fit_transform(x_cat)
x_cat_pre.head()
```

```python
scaler = StandardScaler().set_output(transform="pandas")
```

```python
x_con_pre = scaler.fit_transform(x_con)
x_cat_pre.head()
```

```python
preprocessed = x_con_pre.join(x_cat_pre)
preprocessed
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(preprocessed, y, test_size=0.2)
```
