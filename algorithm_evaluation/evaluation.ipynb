```python
x = df.drop(columns = ["id", "loan_status"])
y = df[["loan_status"]]
```

```python
cat = list(x.columns[x.dtypes=="object"])
con = list(x.columns[x.dtypes!="object"])
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = make_pipeline(
    SimpleImputer(strategy = "median"),
    StandardScaler()
)
```

```python
cat_pipe = make_pipeline(
    SimpleImputer(strategy = "most_frequent"),
    OneHotEncoder(handle_unknown = "ignore", sparse_output = False)
)
```

```python
pre = ColumnTransformer(
    [
        ("num", num_pipe, con),
        ("cat", cat_pipe, cat)
    ]
).set_output(transform = "pandas")
```

```python
xpre = pre.fit_transform(x)
```

```python
xpre.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(xpre, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
import xgboost as xg
```

```python
models = [
    LogisticRegression(),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    GradientBoostingClassifier(),
    xg.XGBClassifier()

]
```

```python
from sklearn.metrics import f1_score
from sklearn.model_selection import cross_val_score
```

```python
def evaluate_single_model(model, xtrain, ytrain, xtest, ytest):

    #Fit the model
    model.fit(xtrain, ytrain)

    #Predict the results for train and test
    ypred_train = model.predict(xtrain)
    ypred_test = model.predict (xtest)

    #calculate f1_macro for train and test
    f1_train = f1_score(ytrain, ypred_train, average = "macro")
    f1_test = f1_score(ytest, ypred_test, average = "macro")

    #5 fold cross validation
    scores = cross_val_score(model, xtrain, ytrain, cv = 5, scoring = "f1_macro", n_jobs = -1)
    f1_cv = scores.mean()

    #create a dictionary for final results
    res = {
        "model_name": model.__class__.__name__,
        "model" : model,
        "f1_train": f1_train,
        "f1_cv" : f1_cv
    }

    return res

```

```python
def algo_evaluation(models : list, xtrain, ytrain, xtest, ytest):

    #Initialize blank list for results
    results = []

    #Apply for loop on models
    for model in models:
        r = evaluate_single_model(model, xtrain, ytrain, xtest, ytest)
        print(r)
        results.append(r)

    #save the results in dataframe
    res_df = pd.DataFrame(results)

    #Sort the results
    sort_df = res_df.sort_values(by = "f1_cv", ascending = False).reset_index(drop = True)

    #Get the best model
    best_model = sort_df.loc[0, "model"]

    return sort_df.round(4), best_model 
```

```python
res_df, best_model = algo_evaluation(models, xtrain, ytrain, xtest, ytest)
```
