```python
import pandas as pd
df= pd.read_csv("Cars93.csv", na_values=["", "NA"], keep_default_na=False)
df.head()
```

```python
y = df[["Weight"]]
```

```python
y.head()
```

```python
x = df.drop(columns = ["id", "Weight"])
```

```python
x.head()
```

```python
x_cat = list(x.columns[x.dtypes == "object"])
x_con = list(x.columns[x.dtypes!="object"])
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = make_pipeline(
    SimpleImputer(strategy="mean"),
    StandardScaler()
)
```

```python
cat_pipe = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OrdinalEncoder()
)
```

```python
pre_obj = ColumnTransformer(
    [
        ("num", num_pipe, x_con),
        ("cat", cat_pipe, x_cat)
    ]
).set_output(transform="pandas")
```

```python
pre_obj
```

```python
processed = pre_obj.fit_transform(x)
```

```python
processed.head()
```

```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LinearRegression
```

```python
model1 = LinearRegression()
```

```python
sel = SequentialFeatureSelector(model1, n_features_to_select="auto", direction="forward")
```

```python
sel
```

```python
sel.fit_transform(processed, y)
```

```python
imp_cols = sel.get_feature_names_out()
```

```python
imp_cols
```

```python
final_x = processed[imp_cols]
```

```python
final_x.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(final_x, y, test_size=0.2)
```

```python
model2 = LinearRegression()
```

```python
model2.fit(xtrain, ytrain)
```

```python
model2.score(xtrain, ytrain)
```

```python
predict = model2.predict(xtest)
predict
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(predict, ytest)
```
