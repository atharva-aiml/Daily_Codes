```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

```

```python
###Create a line plot
```

```python
plt.plot(x, y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Basic line plot")
```

```python
#### Create customized line plot
```

```python
plt.plot(x, y, color = "red", linestyle = '--', marker = 'o', linewidth = 3)
plt.grid(True)
```

```python
#### create multiplots
```

```python
plt.figure(figsize = (9, 5))
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y, color = "green")
plt.title("plot1")

plt.subplot(2, 2, 2)
plt.plot(x, y, color = "green")
plt.title("plot2")

plt.subplot(2, 2, 3)
plt.plot(x, y, color = "green")
plt.title("plot3")
```

```python
#### bar plot

categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 6]

plt.bar(categories, values, color = 'purple')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Basic bar plot")
```

```python
#### create a scatter plot #####
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 4, 5, 6]

plt.scatter(x, y, color = 'blue', marker = 'x')
```

```python
###### pipe chart #####

labels = ['A', 'B', 'C', 'D']
sizes = [30, 20, 40, 10]
colors = ['gold', 'yellowgreen', 'lightgreen', 'skyblue']
explode = (0.2, 0, 0, 0)

plt.pie(sizes, labels= labels, explode = explode)
```
